```python
# cho tờ tiền 100k, hỏi có tối đa bao nhiêu cách đổi ra các tờ mệnh giá nhỏ hơn như 2.000, 3.000 và 5.000
count = 0
for x in range(0,51,1):
    for y in range(0,34,1):
        for z in range(0,21,1):
            if 2*x + 3*y + 5*z == 100:
                print(f'{x} tờ 2000, {y} tờ 3000, {z} tờ 5000')
                count +=1
print(f'Có tất cả {count} cách')
```
