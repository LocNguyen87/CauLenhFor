```python
# Bài 1: 
# Nhập vào một chuỗi gồm các số tự nhiên cách nhau một khoảng trắng. 
# Tính tổng, tính trung bình cộng các số. 
# Tìm số âm lớn nhất và số dương bé nhất trong chuỗi vừa nhập.
# -2 -6 9 -7 8 10 56 -72 -8 -1 6 90 45 21 -5

# chuoi = input('Nhập chuỗi số cách nhau một khoảng trắng: ')
chuoi = '-2 -6 9 -7 8 10 56 -72 -8 -1 6 90 45 21 -5'
number_list = chuoi.split()
tong = max_am = min_duong = 0
so_am = []
so_duong = []
for i in number_list:
    tong += int(i)
    if int(i) < 0:
        so_am.append(i)
    elif int(i) > 0:
        so_duong.append(i)
max_am = so_am[0]
min_duong = so_duong[0]
for sa in so_am:
    if int(sa) > int(max_am):
        max_am = sa
for sd in so_duong:
    if int(sd) < int(min_duong):
        min_duong = sd
trungbinhcong = tong / len(number_list)
print('Chuỗi số:', number_list)
print('Các số âm:', so_am)
print('Các số dương:', so_duong)
print(f'+ Tổng các số trong chuỗi là: {tong}')
print(f'+ Số lượng các số trong chuỗi là: {len(number_list)}')
print(f'+ Trung bình cộng các số trong chuỗi là: {trungbinhcong}')
print(f'+ Số âm lớn nhất là: {max_am}')
print(f'+ Số dương nhỏ nhất là: {min_duong}')
```

```python
import turtle
cw = 40
ch = 40
iniPointX = startPointX = -(cw*4)
iniPointY = startPointY = -(ch*4 + ch/2)
# initialize turcle
t = turtle.Turtle()
turtle.setup(800, 800)
t.color('brown')
t.shape('turtle')
t.pensize(1)
t.fillcolor('brown')
t.speed('fast')
t.penup()
t.goto(startPointX, startPointY)
# draw horizontal lines
h = 0
while h < 10:
    t.pendown()
    t.forward(cw*8)
    t.penup()
    if h < 9:
        startPointY += ch
    t.goto(startPointX, startPointY)
    h+=1
t.right(90)
# draw upper vertical lines
v1 = 0
while v1 < 9:
    t.pendown()
    t.forward(cw*4)
    t.penup()
    if v1< 8:
        startPointX += cw
    t.goto(startPointX, startPointY)
    v1+=1
# draw lower vertical lines
startPointY = 0 - startPointY
t.goto(startPointX, startPointY)
t.left(180)
v2 = 0
while v2 < 9:
    t.pendown()
    t.forward(cw*4)
    t.penup()
    if v2< 8:
        startPointX -= cw
    t.goto(startPointX, startPointY)
    v2+=1
# missing vertical lines
t.forward(ch*4)
t.pendown()
t.forward(ch)
t.penup()
t.right(90)
t.forward(ch*8)
t.right(90)
t.pendown()
t.forward(ch)
t.penup()
# lower diagonal lines
t.goto(iniPointX+(cw*3), iniPointY)
t.left(135)
t.pendown()
t.goto(iniPointX+(cw*5), iniPointY+(ch*2))
t.penup()
t.goto(iniPointX+(cw*3), iniPointY+(ch*2))
t.pendown()
t.goto(iniPointX+(cw*5), iniPointY)
t.penup()
# upper diagonal lines
t.goto(iniPointX+(cw*5), iniPointY+(ch*7))
t.left(90)
t.pendown()
t.goto(iniPointX+(cw*3), iniPointY+(ch*9))
t.penup()
t.goto(iniPointX+(cw*5), iniPointY+(ch*9))
t.left(90)
t.pendown()
t.goto(iniPointX+(cw*3), iniPointY+(ch*7))
t.penup()
# special points
fp_cordinates = [[1,2], [2,3], [4,3], [6,3], [7,2], [1,7], [2,6], [4,6], [6,6], [7,7]]
for fpc in fp_cordinates:
    centerX = iniPointX+(cw*fpc[0])
    centerY = iniPointY+(ch*fpc[1])
    circleX = iniPointX+(cw*fpc[0]-2)
    circleY = iniPointY+(ch*fpc[1]+3)
    ps = [[centerX-cw/8, centerY+ch/8], [centerX+cw/8, centerY+ch/8], [centerX-cw/8, centerY-ch/8], [centerX+cw/8, centerY-ch/8]]
    # small dot
    t.goto(circleX, circleY)
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    # small Ls
    for px, p in enumerate(ps):
        if px == 0:
            t.goto((p[0])-5, p[1])
            t.pendown()
            t.goto((p[0]), p[1])
            t.goto((p[0]), p[1]+5)
            t.penup()
        elif px == 1:
            t.goto((p[0]), p[1]+5)
            t.pendown()
            t.goto((p[0]), p[1])
            t.goto((p[0])+5, p[1])
            t.penup()
        elif px == 2:
            t.goto((p[0]-3), p[1])
            t.pendown()
            t.goto((p[0]), p[1])
            t.goto((p[0]), p[1]-5)
            t.penup()
        elif px == 3:
            t.goto((p[0]+5), p[1])
            t.pendown()
            t.goto((p[0]), p[1])
            t.goto((p[0]), p[1]-5)
            t.penup()
hp_cordinates = [[0,3], [0, 6], [8, 3], [8,6]]
for hpcx, hpc in enumerate(hp_cordinates):
    centerX = iniPointX+(cw*hpc[0])
    centerY = iniPointY+(ch*hpc[1])
    circleX = iniPointX+(cw*hpc[0]-2)
    circleY = iniPointY+(ch*hpc[1]+3)
    # small dot
    t.goto(circleX, circleY)
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.penup()
    # small Ls
    if hpcx == 0 or hpcx == 1:
        t.goto(centerX+cw/8, centerY+ch/8 + 3)
        t.pendown()
        t.goto(centerX+cw/8, centerY+ch/8)
        t.goto(centerX+cw/8 + 5, centerY+ch/8)
        t.penup()
        
        t.goto(centerX+cw/8, centerY-ch/8 - 5)
        t.pendown()
        t.goto(centerX+cw/8, centerY-ch/8)
        t.goto(centerX+cw/8 + 5, centerY-ch/8)
        t.penup()
    else:
        t.goto(centerX-cw/8, centerY+ch/8 + 3)
        t.pendown()
        t.goto(centerX-cw/8, centerY+ch/8)
        t.goto(centerX-cw/8 - 5, centerY+ch/8)
        t.penup()
        
        t.goto(centerX-cw/8 - 5, centerY-ch/8)
        t.pendown()
        t.goto(centerX-cw/8, centerY-ch/8)
        t.goto(centerX-cw/8, centerY-ch/8 - 5)
        t.penup()
# big border
t.pensize(2)
border_bottom_left = [iniPointX - cw/5, iniPointY - ch/5]
border_top_left = [iniPointX - cw/5, iniPointY + (ch*9) + ch/5]
border_top_right = [iniPointX + (cw*8) + cw/5, iniPointY + (ch*9) + ch/5]
border_bottom_right = [iniPointX + (cw*8) + cw/5, iniPointY - ch/5]
t.goto(border_bottom_left[0], border_bottom_left[1])
t.pendown()
t.goto(border_top_left[0], border_top_left[1])
t.goto(border_top_right[0], border_top_right[1])
t.goto(border_bottom_right[0], border_bottom_right[1])
t.goto(border_bottom_left[0], border_bottom_left[1])
# done

t.hideturtle()
turtle.done()
```
